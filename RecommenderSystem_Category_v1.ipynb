```python
reader = Reader(rating_scale=(1,5))
dataset_ecommerce = Dataset.load_from_df(dataset[['customer_unique_id', 'product_category_name', 'review_score_y']], reader)
trainset, testset = train_test_split(dataset_ecommerce, test_size=.25)
sim_options = {'name': 'cosine', 'user_based': False, 'min_support': 6}
algo = KNNWithMeans(sim_options=sim_options)
algo.fit(trainset)
```

```python
predictions = algo.test(testset, verbose = False)
```

```python
def get_Iu(uid):
    try:
        return len(trainset.ur[trainset.to_inner_uid(uid)])
    except ValueError:
        return 0
    
    
def get_Ui(iid):
     try:
        return len(trainset.ir[trainset.to_inner_iid(iid)])
     except ValueError:
        return 0


df = pd.DataFrame(predictions, columns=['uid','iid','rui', 'est', 'details'])
df['Iu'] = df.uid.apply(get_Iu) #number of items reviewed by given user
df['Ui'] = df.iid.apply(get_Ui) #number of users that have reviewed given item
df['err'] = abs(df.est - df.rui)
```
